```python
df = pd.read_csv("https://raw.githubusercontent.com/zaferdemirkol/HerkesIcinYapayZeka/main/reklamlar.csv")
df
```

```python
# Dropping a Column
#  removes a column named "Unnamed: 0" (which might be an automatically generated index column)
#  inplace=True modifies the df itself instead of creating a new one. 
# axis=1 specifies that the operation is done on columns (axis=0 would be for rows).

df.drop("Unnamed: 0",inplace=True, axis=1)
```

```python
df
```

```python
# Checking for missing values
# checks for missing values (like NaN) in each column of df and displays the count of missing values.

df.isnull().sum()
```

```python
df.info()
```

```python
# uses the displot function to create a distribution plot (histogram) of the values in the "TV" column of df. 
# It uses 5 bins to represent the data distribution.

sns.displot(x=df["TV"], bins=5)
```

```python
# counts the occurrences of each unique value in the "ÅŸehir" (city) column.
df["ÅŸehir"].value_counts()
```

```python
# Bar Chart for City Frequency
# As in the previous step plus uses the plot function to create a bar chart showing the frequency of each city.
df["ÅŸehir"].value_counts().plot(kind="bar")
```

```python
# Filtering Data by City
# Filters the df to only include rows where the value in the "ÅŸehir" column is "Ä°stanbul". 
# This creates a new DataFrame containing only data for Istanbul.

df[df["ÅŸehir"]=="Ä°stanbul"]
```

```python
# creates a pivot table from df. It uses "ÅŸehir" as the row index, 
# "tÃ¼r" (category) as the column index, 
# "TV" as the value, and calculates the mean (average) using the aggfunc='mean' argument. 
# The resulting pivot_table shows average TV ad spend for each city and category.

pivot_table = df.pivot_table(index='ÅŸehir', columns='tÃ¼r', values='TV', aggfunc='mean')
pivot_table
```

```python
# Filtering by TV Ad Spend
# filters df to include only rows where the value in the "TV" column is greater than 200. 
# This creates a new DataFrame containing data for companies spending more than 200 on TV ads.
df[df["TV"]>200]
```

```python
# sets the figure size (width and height) and then uses sns.jointplot to create a scatter plot.
# It shows the relationship between "TV" (x-axis) and "satÄ±ÅŸlar" (sales, y-axis).
plt.figure(figsize = (40, 14))
sns.jointplot(x=df["TV"], y=df["satÄ±ÅŸlar"], kind="scatter")
```

```python
# Similar to the previous line, this creates a joint plot but uses the "reg" kind to fit a regression line to the data. 
# This helps visualize the overall trend in sales based on TV ad spend.

sns.jointplot(x=df["TV"], y=df["satÄ±ÅŸlar"], kind="reg")
```

```python
# creates a pair plot that shows the relationship between all numerical columns in df
#  using various plots (scatter, histogram) for each pair.
sns.pairplot(df)
```

```python
#creates a bar chart similar to the one you saw earlier, 
# but it specifies the colors for each bar using the palette argument. The list ["red", "green", "purple"] 
# assigns these colors to the bars representing the three cities in the "ÅŸehir" column.
sns.barplot(x = df['TV'], y = df['ÅŸehir'], palette=["red", "green", "purple"])
```

```python
# Calculating Average TV Ad Spend for Each City:
df[df['ÅŸehir']=="Ä°stanbul"]["TV"].mean()
```

```python
df[df['ÅŸehir']=="Ä°zmir"]["TV"].mean()
```

```python
df[df['ÅŸehir']=="Ankara"]["TV"].mean()
```

```python
#  Counting City Occurrences
# creates a count plot that shows the number of occurrences of each unique value in the "ÅŸehir" column. 
# It's a simple way to visualize the distribution of cities in the data.
sns.countplot(x=df["ÅŸehir"])
```

```python
# Categorical Count Plot (City vs. Category):
# creates a categorical count plot that shows the distribution of "tÃ¼r" (category) for each city. 
# The hue argument separates the counts based on the "tÃ¼r" column, 
# making it easier to compare the frequency of different categories in each city.
sns.catplot(x=df["ÅŸehir"], hue=df["tÃ¼r"], kind="count")
```

```python
# creates a box plot that shows the distribution of the "TV" column. 
# The box plot helps visualize the median, quartiles, and potential outliers in the TV ad spend data.
sns.boxplot(x=df["TV"])
```

```python
# create box plots that show the distribution of "TV" for each city. 
# The first line has "TV" on the x-axis and "ÅŸehir" on the y-axis, while the second line has them reversed. 
# This allows you to compare the TV ad spend distributions across different cities.
sns.boxplot(x=df["TV"], y=df["ÅŸehir"])
```

```python

sns.boxplot(y=df["TV"], x=df["ÅŸehir"])
```

```python
# Creates a violin plot that combines a box plot and a kernel density estimate (KDE) plot. 
# It provides a richer representation of the distribution of "TV" for each city, including the density of data points.
sns.violinplot(y=df["TV"], x=df["ÅŸehir"], palette="magma")
```

```python
# Creates a stripplot that shows individual data points for the "TV" column, categorized by the "ÅŸehir" column.
# Each dot on the plot represents a specific data point.
# palette="magma" argument sets the color scheme for the dots using the "magma" colormap.
sns.stripplot(y=df["TV"], x=df["ÅŸehir"], palette="magma")
```

```python
# Similar to the stripplot, this plot shows individual data points for "TV" categorized by "ÅŸehir".
# However, the swarmplot attempts to avoid overlapping dots by adjusting their positions horizontally.
# This can make it easier to visualize the distribution of data points when there are many overlapping points in the stripplot.
sns.swarmplot(y=df["TV"], x=df["ÅŸehir"], palette="magma")
```

```python
# Facet Grid (Histograms by City and Category)
# creates a facet grid that arranges histograms in a grid based on the "ÅŸehir" and "tÃ¼r" columns.
# Each cell in the grid represents a combination of city and category.
#The map function applies the sns.histplot function to each cell, 
# creating a histogram of the "TV" column for that specific city and category.
# This visualization helps you compare the distribution of TV ad spend across different cities and categories.
sns.FacetGrid(df, col = 'ÅŸehir', row = 'tÃ¼r').map(sns.histplot, "TV")
```

```python
# Linear Regression Plot (Sales vs. TV Ads, Colored by City):
# This line creates a linear regression plot showing the relationship between "TV" (x-axis) and "satÄ±ÅŸlar" (sales, y-axis).
# The hue="ÅŸehir" argument adds a color dimension to the plot, separating the data points based on the "ÅŸehir" column.
# This helps visualize how the relationship between TV ads and sales might differ across different cities.
sns.lmplot(x = 'TV', y = 'satÄ±ÅŸlar', data = df, hue="ÅŸehir")
```
