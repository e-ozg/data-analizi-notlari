## SEABORN

It builds on top of matplotlib, offering a higher-level interface for creating statistical graphics. It provides pre-defined styles, color palettes, and easier-to-use functions for common plots.

Seaborn has built-in datasets, including the 'penguins' dataset, which you can access directly within your code. This dataset contains information about different penguin species.

```python
import seaborn as sns
```

```python
df = sns.load_dataset('penguins')  # Loads the penguins dataset
df.head() # Shows the first few rows
```

```python
df.tail() # Shows the last few rows
```

```python
df.shape # Shows the number of rows and columns
```

```python
df.isnull().sum() # Shows the number of missing values in each column
```

```python
df.dropna(how='any', inplace=True) # Removes rows with any missing values (be cautious!)
```

```python
df.isnull().sum()
```

```python
df.shape
```

```python
df.info() # Displays data types and non-null counts
```

```python
df.describe().T # Provides descriptive statistics (mean, std, etc.)
```

```python
sns.displot(x=df['species']) # Distribution plot of 'species'
```

```python
sns.countplot(x=df['species']) # Count plot of 'species' (number of penguins in each species)
plt.title('Penguen turleri') # Sets title
plt.show()
```

```python
sns.countplot(x=df['species'], stat='percent') # Count plot with percentages
plt.title('Penguen turleri')
plt.show()
```

```python
sns.countplot(x=df['sex']) # Count plot of 'sex'
plt.title('Cinsiyet sayilari') # Sets title
plt.show()
```

```python
df['sex'].value_counts().plot(kind='bar') # Alternative count plot using Series methods
```

```python
sns.catplot(x=df['species'], hue=df['island'], kind='count') # Countplot with species and island categories
```

```python
sns.catplot(hue=df['species'], x=df['island'], kind='count') # Reverse category order
```

```python
df.head()
```

```python
df['island'].unique() # Shows unique island names
```

```python
df['island'].value_counts() # Shows counts of each island
```

```python
df['island'].value_counts(normalize=True) # Shows normalized counts (percentages)
```

```python
sns.countplot(x='island', hue='species', data=df) # Specify data explicitly
```

```python
sns.countplot(hue='island', x='species', data=df) # Reverse category order
```

```python
df.groupby(['island', 'species']) # Group by island and species
```

```python
df.groupby(['island', 'species']).agg({'bill_length_mm': ['mean','min','max']}) # Aggregate bill_length_mm for each group
```

```python
# Pivot table for calculating mean body mass per island and species:
pt = df.pivot_table(
    index='island',
    columns='species',
    values='body_mass_g',
    aggfunc='mean'
)
pt # View the pivot table
```

```python
pt.plot(kind='bar', figsize=(10,6)) # Create a bar chart from the pivot table
plt.title('Tur - ada bazinda agirlik ortalamalari') # Title: Species-Island Weight Averages
plt.xlabel('Ada') # X-axis label: Island
plt.ylabel('Agirlik') # Y-axis label: Weight
plt.show() # Display the plot
```

```python
sns.jointplot(x='body_mass_g', y='flipper_length_mm', data=df) # Create a joint distribution plot
```

```python
# Display the joint distribution plot with a regression line
sns.jointplot(x='body_mass_g', y='flipper_length_mm', data=df, kind='reg')
```

```python
df.head()
```

```python
# Create a linear regression model with color-coded points by species
sns.lmplot(x='body_mass_g', y='flipper_length_mm', data=df, hue='species')
```

```python
# Create a pair plot showing relationships between all numerical features
sns.pairplot(df)
```

```python
# Create a pair plot showing relationships between all numerical features, colored by species
sns.pairplot(df, hue='species')
```
