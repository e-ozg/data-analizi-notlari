```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
punk = pd.read_csv("punk.csv")
punk
```

```python
# Aykırı değerleri çıkartan fonksiyon

def aykiri_degerleri_cikart(df, sutun_adi, m=2):
  ort = df[sutun_adi].mean()
  std = df[sutun_adi].std()
  kesme_degeri = std * m
  alt_sinir, ust_sinir = ort - kesme_degeri, ort + kesme_degeri
  filtrelenmis = df[ (df[sutun_adi] >= alt_sinir) & (df[sutun_adi] <= ust_sinir) ]
  return filtrelenmis
```

```python
df_aykirisiz = aykiri_degerleri_cikart(punk, 'duration_seconds')
df_aykirisiz
```

```python
punk_rating = aykiri_degerleri_cikart(punk, 'rating')
punk_rating
```

```python
def eksik_ortalamasi(df):
    percent_nan = 100 * df.isnull().sum() / len(df)
    percent_nan = percent_nan[percent_nan > 0].sort_values()
    return percent_nan
percent_nan = eksik_ortalamasi(punk)
percent_nan
```

```python
# Eksik veri yüzdelerini çubuk grafikle görselleştirme
plt.figure(figsize=(8, 5))
sns.barplot(x=percent_nan.index, y=percent_nan)
plt.title("Eksik Veri Yüzdeleri")
plt.show()
```

```python
# Eksik rating değerlerini ortalama ile doldurma

punk['rating'] = punk['rating'].fillna(punk['rating'].mean())
```

```python
# Aykırı değerleri IQR yöntemi ile çıkarma

def aykiri_degerler_iqr(df, sutun_adi):
    Q1 = df[sutun_adi].quantile(0.25)  # %25'lik çeyreklik değer
    Q3 = df[sutun_adi].quantile(0.75)  # %75'lik çeyreklik değer
    IQR = Q3 - Q1  # IQR hesaplama
    alt_sinir = Q1 - 1.5 * IQR  # Alt sınır
    ust_sinir = Q3 + 1.5 * IQR  # Üst sınır
    filtrelenmis = df[(df[sutun_adi] >= alt_sinir) & (df[sutun_adi] <= ust_sinir)]  # Sınırlar dışındaki verileri çıkar
    return filtrelenmis  # Aykırı değerler olmadan DataFrame döndür

punk = aykiri_degerler_iqr(punk, 'duration_seconds')
punk
```

```python
# release_date alanındaki eksik verileri Unknown olarak doldurma

punk['release_date'] = punk['release_date'].fillna('Unknown')
punk.head(10)
```

```python
# Artist sütunundaki verileri Anonymous ile doldurma

punk['artist'] = punk['artist'].fillna('anonymous')
punk.head(15)
```

```python
# Country sütunundaki eksik verileri Unknown ile doldurma

punk['country'] = punk['country'].fillna('Unknown')
punk.head(10)
```
